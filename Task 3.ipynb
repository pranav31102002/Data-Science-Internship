```python
# Problem 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = 0

for name in employees:
    if employees[name] > highest_score:
        highest_score = employees[name]

# Find all employees having highest score
top_performers = []

for name in employees:
    if employees[name] == highest_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", highest_score, ")")
```

```python
# Problem 2: Search Query Keyword Analysis

query = "Buy mobile phone buy phone online"

# convert to lowercase
query = query.lower()

# split into words
words = query.split()

freq = {}

# count frequency
for word in words:
    if word in freq:
        freq[word] = freq[word] + 1
    else:
        freq[word] = 1

# print words searched more than once
result = {}

for word in freq:
    if freq[word] > 1:
        result[word] = freq[word]

print(result)
```

```python
# Problem 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

i = 0

# check each reading
for value in sensor_readings:
    if value % 2 == 0:
        valid_readings.append((i, value))
    i = i + 1

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# count domains
for email in emails:
    parts = email.split("@")
    domain = parts[1]

    if domain in domain_count:
        domain_count[domain] = domain_count[domain] + 1
    else:
        domain_count[domain] = 1

total = len(emails)

# print percentage
for domain in domain_count:
    percentage = (domain_count[domain] / total) * 100
    print(domain + ":", int(percentage), "%")
```
