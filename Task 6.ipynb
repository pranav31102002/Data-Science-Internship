```python
# Smart Parking Lot Management System

def parking_system(capacity, logs):

    parked = 0        # current vehicles
    peak = 0          # highest usage

    # go through each log
    for entry in logs:

        if entry == "IN":
            parked = parked + 1
        elif entry == "OUT":
            parked = parked - 1

        # track peak usage
        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    # check capacity
    if parked > capacity:
        print("Parking Status: Full / Alert!")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_system(capacity, vehicle_logs)
```

```python
# Problem 2

def delivery_time(distance, traffic, weather):

    # base time = 5 min per km
    time = distance * 5

    # traffic delay
    if traffic == "High":
        time = time + 10
    elif traffic == "Medium":
        time = time + 5

    # weather delay
    if weather == "Rainy":
        time = time + 5
    elif weather == "Storm":
        time = time + 15

    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
delivery_time(8, "High", "Rainy")
```

```python
# Problem 3

def theatre_occupancy(total_seats, booked_list):

    booked = len(booked_list)

    occupancy = (booked / total_seats) * 100
    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input (150 bookings)
booked = [1] * 150
theatre_occupancy(200, booked)
```

```python
# Problem 4

def server_load(cpu_list):

    total = 0

    for cpu in cpu_list:
        total = total + cpu

    avg = total / len(cpu_list)

    print("Average CPU Load:", int(avg), "%")

    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
server_load([45, 60, 70, 85, 90])
```
