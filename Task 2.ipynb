```python
# Problem 1: Unique Words in a Sentence

# Take Input
sentence = input("Enter sentence: ")
# Split sentence into words
words = sentence.split()
# Convert list into set to remove duplicates
unique_words = set(words)

# Output
print("Unique words count :- ", len(unique_words))
print("Unique words :- ", unique_words)
```

```python
# Problem 2: Highest Salary from Employee Data

# Employee dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Finding highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
# Problem 3: Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]

# Initialize max and min
max_val = numbers[0]
min_val = numbers[0]

# Loop through list
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

```python
# Problem 4: Count Products Above a Price Threshold

prices = [450, 1200, 899, 1500, 300]

count = 0

# Check prices greater than 1000
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
# Problem 5: Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

# Count Present days
present_days = attendance.count("P")

# Calculate percentage
percentage = (present_days / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

```python
# Problem 6: Remove Duplicate Phone Numbers

phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set
unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

```python
# Problem 7: Count Character Frequency

text = "python"

char_count = {}

# Count each character
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

print(char_count)
```

```python
# Problem 8: Convert List to Tuple

my_list = [10, 20, 30]

my_tuple = tuple(my_list)

print("Tuple:", my_tuple)
```

```python
# Problem 9: Check if a Key Exists in a Dictionary

employees = {"Ravi": 75000, "Anita": 68000}

key = "Ravi"

if key in employees:
    print("Employee exists")
else:
    print("Employee not found")
```

```python
# Problem 10: Calculate Average Marks

marks = [70, 80, 65, 75]

# Calculate average
average = sum(marks) / len(marks)

print("Average Marks:", average)
```
