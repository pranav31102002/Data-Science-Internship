```python
# Problem 1

def analyze_engagement(likes_list):
    total_likes = 0

    # calculate total likes
    for like in likes_list:
        total_likes = total_likes + like

    print("Total Likes:", total_likes)

    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


likes = [200, 300, 250, 300]
analyze_engagement(likes)
```

```python
# Problem 2

def medicine_stock(stock):
    print("Medicine Stock:", stock)

    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")


medicine_stock(6)
```

```python
# Problem 3

def rainfall_check(rainfall, required):
    total = 0

    for r in rainfall:
        total = total + r

    avg = total / len(rainfall)

    print("Average Rainfall:", int(avg))

    if avg >= required:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


data = [60, 70, 80, 78]
rainfall_check(data, 70)
```

```python
# Problem 4

def check_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


users = ["ravi", "anita", "ravi", "kiran"]
check_duplicates(users)
```

```python
# Problem 6

def premium_crops(prices):
    premium_list = []

    for price in prices:
        if price > 2000:
            premium_list.append(price)

    print("Premium Crops:", premium_list)


crop_prices = [1500, 2500, 3200, 1800]
premium_crops(crop_prices)
```

```python
# Problem 7

def app_health(errors):
    print("Error Count:", errors)

    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


app_health(7)
```

```python
# Problem 8

def transaction_check(amount):
    print("Transaction Amount:", amount)

    if amount <= 50000:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


transaction_check(60000)
```

```python
# Problem 9

def attendance_check(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present = present + 1

    percentage = (present / len(attendance)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


att = ["P", "P", "A", "P", "P"]
attendance_check(att)
```
