```python
# Problem 1: Mobile Recharge Validation System

# function to validate recharge
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50")
        return False
    elif amount not in valid_plans:
        print("Invalid plan selected")
        return False
    else:
        print("Recharge Successful!")
        return True


# retry using while loop
while True:
    amt = int(input("Enter recharge amount: "))
    if validate_recharge(amt):
        break
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    for item in products:
        if products[item] < 15:
            print(item, "→ Reorder Alert")
        else:
            print(item, "→ Stock OK")


inventory = {
    "Keyboard": 10,
    "Mouse": 25,
    "Monitor": 12,
    "USB Cable": 30
}

check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def check_result(marks):
    total = 0

    for m in marks:
        total = total + m

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


marks_list = [60, 40, 70, 55]
result = check_result(marks_list)
print(result)
```

```python
# Problem 4: Cab Fare Estimator

def calculate_fare(km, peak):
    fare = 50 + (12 * km)

    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare

while True:
    distance = int(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak_hour)
    print("Total Fare:", total_fare)

    retry = input("Calculate again? (yes/no): ")
    if retry == "no":
        break
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def attendance_check(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present = present + 1

    percentage = (present / len(attendance)) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

att_list = ["P", "P", "A", "P", "P"]
print(attendance_check(att_list))
```

```python
# Problem 6: Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False

    if len(password) < 8:
        print("Weak Password: Minimum length 8 required")
        return

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if has_digit and has_special:
        print("Strong Password")
    else:
        print("Weak Password")

pwd = input("Enter password: ")
check_password(pwd)
```
