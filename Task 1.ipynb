```python
# 1. User Login Check
user_name = "admin"
password = "1234"

input_username = input("Enter username: ")
input_password = input("Enter password: ")

if input_username == user_name and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# Problem 2: Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count = pass_count + 1
    else:
        fail_count = fail_count + 1

print("Total Students:", len(marks))
print("Pass:", pass_count)
print("Fail:", fail_count)
```

```python
# Problem 3: Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
new_names = []
for name in names:
    name = name.strip()
    name = name.lower()
    new_names.append(name)
print(new_names)
```

```python
# Problem 4: Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print("Message:", msg)
    print("Length:", length)
    
    if length > 10:
        print("Long message")
```

```python
# Problem 5: Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count = 0
for log in logs:
    if log == "ERROR":
        count = count + 1

print("Total ERROR:", count)
```
